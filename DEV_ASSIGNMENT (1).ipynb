```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import plotly.express as px

df1=pd.read_csv('/content/winequalityN.csv')

df=df1.head(50)
df
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df["fixed acidity"],df["quality"])
plt.xlabel('Acidity')
plt.ylabel('Quality')
plt.title('SCATTER PLOT')
```

```python
plt.figure(figsize=(10,10))
plt.bar(df["fixed acidity"],df["volatile acidity"])
plt.xlabel('fixed acidity')
plt.ylabel('volatile acidity')
plt.title('BAR GRAPH')
```

```python
plt.figure(figsize=(10,10))
plt.plot(df["fixed acidity"],df['chlorides'])
plt.xlabel('fixed acidity')
plt.ylabel('chlorides')
plt.title('SIMPLE LINE PLOT')
```

```python
df2=df1.head(10)
plt.figure(figsize=(10,10))
label=df2['quality']
x=df2['fixed acidity']
plt.pie(x,labels=label)
plt.title('PIE CHART')
```

```python
fig, axes = plt.subplots()
  
# plotting using plt.stem
axes.stem(df["fixed acidity"],df["quality"],
          use_line_collection=True, basefmt=' ')
  
# starting value of y-axis
axes.set_ylim(0)
  
# details and formatting of chart
plt.title('LOLLIPOP CHART')
plt.xlabel('fixed acidity')
plt.ylabel('quality')
```
